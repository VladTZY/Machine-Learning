# Logistic Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
import dtreeviz
from sklearn import tree


data = pd.read_csv("diabetes.csv")

dx = data.drop("Outcome", axis = 1)
dy = data[["Outcome"]]

repeats = 30
accuracies = []

for i in range(0, repeats):
    dx_train, dx_test, dy_train, dy_test = train_test_split(dx, dy, test_size = 0.25)

    dtm = DecisionTreeClassifier()
    dtm.fit(dx_train, dy_train)

    dy_pred = dtm.predict(dx_test)
    accuracies.append(accuracy_score(dy_test, dy_pred))

print("Mean accuracy", np.mean(accuracies))
print("Standard deviation", np.std(accuracies))
confusion_matrix(dy_test, dy_pred)

plt.figure(figsize=(15, 15))
tree.plot_tree(dtm, rounded= True, filled=True, proportion=True)
```
